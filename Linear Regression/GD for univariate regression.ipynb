```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Load and preprocess the data
df = pd.read_csv("cleanedHouseData.csv")
df = df.head(30)
x_train = df["LotArea"] / 1000  # Normalize the input features
y_train = df["Price"] / 1000  # Normalize the target variable

# Define the linear regression model
def predict(x, w, b):
    return x * w + b

# Generate and plot the model's predictions
def plot_predictions(x, y, w, b):
    plt.scatter(x, y, label="Actual Data")
    f_wb = predict(x, w, b)
    plt.plot(x, f_wb, label="Linear Regression Model", color='red')
    plt.xlabel("Lot Area (in 1000 sq ft)")
    plt.ylabel("Price (in 1000 dollars)")
    plt.legend()
    plt.show()

# Train the model using gradient descent
def gradient_descent(x, y, w, b, alpha=0.001, num_iter=10000):
    m = x.shape[0]
    for i in range(num_iter):
        d_dw, d_db = derivative(x, y, w, b)
        w -= alpha * d_dw
        b -= alpha * d_db
    return w, b

# Compute the cost function
def compute_cost(x, y, w, b):
    m = x.shape[0]
    cost = np.sum((predict(x, w, b) - y) ** 2)
    total_cost = cost / (2 * m)
    return total_cost

# Compute derivatives
def derivative(x, y, w, b):
    m = x.shape[0]
    sum_w = np.sum((predict(x, w, b) - y) * x)
    sum_b = np.sum(predict(x, w, b) - y)
    return sum_w / m, sum_b / m

# Initialize model parameters to zero
w = 0.0
b = 0.0

# Train the model
w, b = gradient_descent(x_train, y_train, w, b, alpha=0.0001, num_iter=100000)

# Plot the final model predictions
plot_predictions(x_train, y_train, w, b)

# Make a prediction for a new input
x_i = float(input("Enter the house area (in 1000 sq ft): "))
y_hat = predict(x_i, w, b)

print(f"The predicted price is {y_hat * 1000} dollars")
print(f"The cost is {compute_cost(x_train, y_train, w, b)}")
```
